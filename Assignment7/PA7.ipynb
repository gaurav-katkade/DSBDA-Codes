```python
import pandas as pd
import numpy as np
```

```python
text=open('Pune.txt').read()
```

```python
text
```

```python
import nltk
```

```python
from nltk import sent_tokenize
from nltk import word_tokenize
nltk.download('punkt')
```

```python
sent_tokens=sent_tokenize(text)
```

```python
sent_tokens
```

```python
word_tokens=word_tokenize(text)
```

```python
word_tokens
```

```python
from nltk.stem import PorterStemmer
s=PorterStemmer()
stem_array=[]
for i in word_tokens:
    j=s.stem(i)
    stem_array.append(j)
```

```python
stem_array
```

```python
nltk.download('averaged_perceptron_tagger')
pos_tags=nltk.pos_tag(word_tokens)
```

```python
pos_tags
```

```python
from nltk.corpus import stopwords
```

```python
nltk.download('stopwords')
sw=stopwords.words('english')
sw
```

```python
without_sw=[]
for i in word_tokens:
    if i not in sw:
        without_sw.append(i)
```

```python
without_sw
```

```python
from nltk.stem import WordNetLemmatizer
nltk.download('wordnet')

lemmetizer=WordNetLemmatizer()
pos_map={
    'J':'a','V':'v','N':'n','R':'r'
}

lem=[]
for word,pos in pos_tags:
    ptw=pos_map.get(pos[0].upper(),'n')
    lemm_word=lemmetizer.lemmatize(word,ptw)
    lem.append(lemm_word)
```

```python
lem
```
